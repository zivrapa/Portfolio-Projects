```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import codecademylib3_seaborn
import glob

files = glob.glob('states*.csv') # load all the files of the cencus

df_list = []
for item in files:
    data = pd.read_csv(item)
    df_list.append(data) # LOOP OVER THE FILES AND APPEND THEM TO ONE LIST

df = pd.concat(df_list) # concat them into one dataframe

# print(df.Income.max())

print(df.columns)


df['Income'] = df['Income'].replace({'\$':''}, regex = True)
df['Income'] = pd.to_numeric(df['Income'])

# taking off the $ sigh and turn the column to numeric type 

df[['Men', 'Women']] = df['GenderPop'].str.split('_', expand=True)

# men and women pop. display together - so I split them into 2 columns and turn it to numeric type 

df['Men'] = df['Men'].replace({'\D':''}, regex = True)
df['Women'] = df['Women'].replace({'\D':''}, regex = True)
df['Men'] = pd.to_numeric(df['Men'])
df['Women'] = pd.to_numeric(df['Women'])
df.set_index('State', inplace = True)
df = df.drop(['Unnamed: 0', 'GenderPop'], axis = 1)
print(df.head())

df.loc['Maryland']['Women'] = df.loc['Maryland']['Women'].fillna(value = (df.loc['Maryland']['TotalPop']- df.loc['Maryland']['Men']))

df['Women'] = df['Women'].fillna(value = (df.loc['Montana']['TotalPop']- df.loc['Montana']['Men']))
df.duplicated()
df = df.drop_duplicates()
print(df)
plt.scatter(df['Women'], df['Men']) 
# plt.show()

# fill empty cells in the data frame - calulate the women pop. by subtract the men pop. from the total pop.

df['Hispanic'] = df['Hispanic'].replace({'\%':''}, regex = True)
df['Hispanic'] = pd.to_numeric(df['Hispanic'])
df['White'] = df['White'].replace({'\%':''}, regex = True)
df['White'] = pd.to_numeric(df['White'])
df['Black'] = df['Black'].replace({'\%':''}, regex = True)
df['Black'] = pd.to_numeric(df['Black'])
df['Native'] = df['Native'].replace({'\%':''}, regex = True)
df['Native'] = pd.to_numeric(df['Native'])
df['Asian'] = df['Asian'].replace({'\%':''}, regex = True)
df['Asian'] = pd.to_numeric(df['Asian'])
df['Pacific'] = df['Pacific'].replace({'\%':''}, regex = True)
df['Pacific'] = pd.to_numeric(df['Pacific'])
print(df.dtypes)

# turn to numeric all the cell that contain data about races.
```
