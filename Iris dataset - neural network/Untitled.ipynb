```python
import tensorflow as tf
import matplotlib.pyplot as plt
plt.rc('figure', autolayout=True)
plt.rc('axes', labelweight='bold', labelsize='large',
       titleweight='bold', titlesize=18, titlepad=10)
plt.rc('animation', html='html5')
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('C:/Users/lab_p/Documents/GitHub/Portfolio-Projects/Iris dataset - neural network/IRIS.csv')
```

```python
X = df[['sepal_length','sepal_width','petal_length','petal_width']].copy()
y = OneHotEncoder().fit_transform(df[['species']].copy()).toarray()
scalar = StandardScaler()
```

```python
# y.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.75)

X_train = scalar.fit_transform(X_train)
X_valid = scalar.transform(X_valid)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import callbacks

model1 = keras.Sequential([
    layers.Dense(4, activation='relu',input_shape=[4]),
    
    layers.Dense(8, activation='relu'),
    
    layers.Dense(8, activation='relu'),
    
    layers.Dense(3, activation='softmax'),
])
```

```python
model1.compile(
    optimizer='adam',
    loss=tf.keras.losses.CategoricalCrossentropy(),
    metrics=[tf.keras.metrics.CategoricalAccuracy()]
)
```

```python
early_stopping = callbacks.EarlyStopping(patience = 15, 
                                        min_delta = 0.1,
                                        restore_best_weights = True)
```

```python
history1 = model1.fit(
    X_train, y_train,
    validation_data=(X_valid, y_valid),
    batch_size=10,
    epochs=350,
    verbose=0, # suppress output since we'll plot the curves
    callbacks = [early_stopping]
)
```

```python
history1_df = pd.DataFrame(history1.history)
```

```python
history1_df
```

```python

history1_df.loc[0:, ['categorical_accuracy', 'val_categorical_accuracy']].plot()
print("Maximum Validation accuracy: {:0.4f}".format(history1_df['val_categorical_accuracy'].max()));
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import callbacks

model = keras.Sequential([
    layers.Dense(4, activation='relu',input_shape=[4]),
    
    layers.Dense(8, activation='relu'),
    
    layers.Dense(8, activation='relu'),
    
    layers.Dense(3, activation='sigmoid'),
])
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.CategoricalCrossentropy(),
    metrics=[tf.keras.metrics.CategoricalAccuracy()]
)
```

```python
early_stopping = callbacks.EarlyStopping(patience = 15, 
                                        min_delta = 0.1,
                                        restore_best_weights = True)
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_valid, y_valid),
    batch_size=10,
    epochs=350,
    verbose=0, # suppress output since we'll plot the curves
    callbacks = [early_stopping]
)
```

```python
history_df = pd.DataFrame(history.history)
```
