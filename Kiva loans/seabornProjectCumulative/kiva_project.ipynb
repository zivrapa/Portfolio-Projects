# Visualizing Loans Awarded by Kiva

In this project we will visualize insights using a dataset from <a href = "https://www.kaggle.com/fkosmowski/kivadhsv1" target = "_blank">Kaggle</a>. The dataset contains information about loans awarded by the non-profit <a href = "https://www.kiva.org/" target = "_blank">Kiva</a>. 

Using Seaborn, we will explore the average loan amount by country using aggregated bar charts. we will also visualize the distribution of loan amounts by project type and gender using box plots and violin plots.


## Step 1: Importing Necessary Python Modules

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

## Step 2: load The Data

```python
df = pd.read_csv('kiva_data.csv ')
```

## Step 3: Examine The Data


### Overview of the dataset:

Each entry (row) in the dataset represents a loan that Kiva awarded to a particular project. The `loan_amount` column shows the amount (in U.S. dollars) awarded to the project. The `activity` column has the category type that the project falls under. The `country` column is the country where the project is located. The `gender` column represents the gender of the primary person who applied for the loan. 


let's print the first 25 rows of `df` using `.head()`

```python
print(df.head(25))
```

## Step 4: Bar Charts

now, let's create a bar plot using Seaborn to visualize the average size of Kiva loans given to projects, by country.

We've set up the figure you'll use to plot your bar plot on.

```python
# Creates the figure
f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(data=df, x='country',y='loan_amount')
```

Adding '$' sign to the ticks of 'y' axis:

```python
import matplotlib.ticker as mtick

# Creates the figure
f, ax = plt.subplots(figsize=(15, 10))

# Plot the data
sns.barplot(data=df, x="country", y = "loan_amount")


fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

```

## Step 5: Learn More By Using `hue` according to gender

```python
# Creates the figure
f, ax = plt.subplots(figsize=(15, 10))

sns.barplot(data=df, x="country", y="loan_amount",hue='gender')

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

On average, male recipients receive larger loans from Kiva.

Pakistan is the country that has the *least* disparity in loan amounts awarded by gender?

## Step 6: Styling


Set a different color palette and different background.

```python
# Set color palette
sns.set_palette('Set2')

# Set style
sns.set_style('white')

# Create figure and axes (no need to use the previous syntax, as the y-label ticks aren't going to be formatted)
plt.figure(figsize=(25, 15))

# Add a title
plt.title('Kiva loans analysis')

# Use Seaborn to create the bar plot
sns.barplot(data=df, x="country", y = "loan_amount")
```

## Step 7: Box Plots With Kiva Data

So far we have visualized the average size of loans by country using bar charts; now we are going to make a box plot to compare the distribution of loans by country.

```python
plt.figure(figsize=(16, 10))

sns.boxplot(data=df, x="country", y = "loan_amount")
```

#### Kenya's box has the widest distribution

#### Cambodia is the country would you be most likely to receive the largest loan amount.

## Step 8: Box Plot by Activity

Instead of visualizing the loan amount by *country*, we use seaborn boxplot to plot the loan amount by *activity*.

```python
plt.figure(figsize=(16, 10))

sns.boxplot(data=df,x='activity',y = "loan_amount")
sns.set_theme(context='poster', palette='bright')
```

#### Farming activity is the most likly to recive the largest loan amount

## Step 9: Violin Plots

#### Violin plot - loan amount by activity 

```python
plt.figure(figsize=(16, 10))

sns.violinplot(data=df, x="activity", y="loan_amount")
```

###  Distribution of loan amount by country.
Previously, we created a violin plot and plotted the data by _activity_. This time, we will create a violin plot that plots the data by _country_.

```python
plt.figure(figsize=(16, 10))
sns.violinplot(data=df, x="country", y="loan_amount")
```

## Step 10: Split Violin Plots

We will use the `hue` and `split` parameters with `sns.violinplot()` to visualize the distribution of loan amount by country, split by gender. 

```python
# Some styling
sns.set_palette("Spectral")
plt.figure(figsize=(18, 12))
sns.violinplot(data=df, x="country", y="loan_amount",hue='gender')
```

What does this visualization reveal about the distribution of loan amounts within countries by gender?

* The loans for men is much more wide-spread 
* In Cambodia and Pakistan there is the biggest gap between men and women
