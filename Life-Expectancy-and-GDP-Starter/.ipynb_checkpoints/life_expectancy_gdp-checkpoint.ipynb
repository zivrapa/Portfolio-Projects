```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

## What is GDP?

The GDP is the total of all value added created in an economy. The value added means the value of goods and services that have been produced minus the value of the goods and services needed to produce them

## what is Life expectency?

Life expectancy is the median age at death for a particular population group. For example, if a group of people have a life expectancy at birth of 70 years, half died before 70 and half survived beyond this age.

```python
data = pd.read_csv('all_data.csv')
data.head(16)
# len(data)
```

```python
data.dtypes
```

```python
labels = data['Country'].unique()
```

```python
chlile_data = data[data.Country == 'Chile']
china_data = data[data.Country == 'China']
germany_data = data[data.Country == 'Germany']
Mexico_data = data[data.Country == 'Mexico']
US_data = data[data.Country == 'United States of America']
Zimbabwe_data = data[data.Country == 'Zimbabwe']

plt.plot(chlile_data.Year,chlile_data['Life expectancy at birth (years)'])
plt.plot(china_data.Year,china_data['Life expectancy at birth (years)'])
plt.plot(germany_data.Year,germany_data['Life expectancy at birth (years)'])
plt.plot(Mexico_data.Year,Mexico_data['Life expectancy at birth (years)'])
plt.plot(US_data.Year,US_data['Life expectancy at birth (years)'])
plt.plot(Zimbabwe_data.Year,Zimbabwe_data['Life expectancy at birth (years)'])
plt.legend(labels)
plt.show()
```

```python
plt.clf()
sns.set()
fig, axes = plt.subplots(2, 3, figsize=(20,8))
fig.subplots_adjust(hspace=0.4)

axes1 = sns.regplot(x='GDP',y='Life expectancy at birth (years)',data=Zimbabwe_data,line_kws = {'color':'black'},ax=axes[0,0])
axes1.title.set_text('GDP vs life expectency, Zimbabwe')
axes1.set_xlabel('GDP (in  ten Billions $)')
axes2 = sns.lineplot(x=Zimbabwe_data.Year,y=Zimbabwe_data['Life expectancy at birth (years)'],data=Zimbabwe_data,ax=axes[0,1])
axes2.title.set_text('life expectency by Years, Zimbabwe')
axes3 = sns.lineplot(x=Zimbabwe_data.Year,y=Zimbabwe_data['GDP'],data=Zimbabwe_data,ax=axes[0,2])
axes3.title.set_text('GDP by Years, Zimbabwe')
axes3.set_ylabel('GDP (in ten billions) $')

axes4 = sns.regplot(x='GDP',y='Life expectancy at birth (years)',data=germany_data,line_kws = {'color':'black'},ax=axes[1,0])
axes4.title.set_text('GDP vs life expectency, Germany')
axes4.set_xlabel('GDP (in Trillions $)')
axes5 = sns.lineplot(x=germany_data['Year'],y=germany_data['Life expectancy at birth (years)'],data=germany_data,ax=axes[1,1])
axes5.title.set_text('life expectency by Years, Germany')
axes6 = sns.lineplot(x=germany_data['Year'],y=germany_data['GDP'],data=germany_data,ax=axes[1,2])
axes6.title.set_text('GDP by Years, Germany')
axes6.set_ylabel('GDP (in Trillions $)')
```

We can see a pretty good corrolation in between the increasing GDP and the growing life expectency.

```python
corr_GDP_life_Zim, p = pearsonr(Zimbabwe_data['GDP'], Zimbabwe_data['Life expectancy at birth (years)'])
print('Zimbabuwe correlation between increase in GDP and life expectency: ' + str(corr_GDP_life_Zim))

corr_GDP_life_Ger, p = pearsonr(germany_data['GDP'], germany_data['Life expectancy at birth (years)'])
print('Zimbabuwe correlation between increase in GDP and life expectency: ' + str(corr_GDP_life_Ger))
```

```python
plt.clf()
sns.set()

fig, axes = plt.subplots(2, 3, figsize=(20,8))
fig.subplots_adjust(hspace=0.4)

ax1 = sns.lineplot(x=chlile_data.Year,y=chlile_data['GDP'],data=chlile_data,ax=axes[0,0]).set(title='Chile')
ax2 = sns.lineplot(x=china_data.Year,y=china_data['GDP'],data=china_data,ax=axes[0,1]).set(title='China')
ax3 = sns.lineplot(x=germany_data.Year,y=germany_data['GDP'],data=germany_data,ax=axes[0,2]).set(title='Germany')
ax4 = sns.lineplot(x=Mexico_data.Year,y=Mexico_data['GDP'],data=Mexico_data,ax=axes[1,0]).set(title='Mexico')
ax5 = sns.lineplot(x=US_data.Year,y=US_data['GDP'],data=US_data,ax=axes[1,1]).set(title='US')
ax6 = sns.lineplot(x=Zimbabwe_data.Year,y=Zimbabwe_data['GDP'],data=Zimbabwe_data,ax=axes[1,2]).set(title='Zimbabwe')
```
